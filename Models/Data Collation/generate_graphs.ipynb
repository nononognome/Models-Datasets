```python
import pandas as pd
from matplotlib import pyplot as plt
import tikzplotlib
import numpy as np
import os
```

# ANN Peak F1, AUC-ROC

```python
max_ann = pd.read_csv(f"highest_ANN.csv")
```

```python
independent = 'Test F1'
variable = 'n_fft'
second_condition = (max_ann[variable] != 0)
scale = 100

colours = {'mel': 'red', 'mfcc': 'blue', 'cqt': 'green', 'spectrogram': 'black'}

mel_test = max_ann[(max_ann['type'] == 'mel') & second_condition][independent] * scale; mel_x = max_ann[(max_ann['type'] == 'mel') & second_condition][variable]
mfcc_test = max_ann[(max_ann['type'] == 'mfcc') & second_condition][independent] * scale; mfcc_x = max_ann[(max_ann['type'] == 'mfcc') & second_condition][variable]
cqt_test = max_ann[(max_ann['type'] == 'cqt') & second_condition][independent] * scale; cqt_x = max_ann[(max_ann['type'] == 'cqt') & second_condition][variable]
spec_test = max_ann[(max_ann['type'] == 'spectrogram') & second_condition][independent] * scale; spec_x = max_ann[(max_ann['type'] == 'spectrogram') & second_condition][variable]

mean_mel = np.array([[x, np.mean(mel_test[mel_x == x])] for x in np.sort(mel_x.unique())])
mean_mfcc = np.array([[x, np.mean(mfcc_test[mfcc_x == x])] for x in np.sort(mfcc_x.unique())])
mean_cqt = np.array([[x, np.mean(cqt_test[cqt_x == x])] for x in np.sort(cqt_x.unique())])
mean_spec = np.array([[x, np.mean(spec_test[spec_x == x])] for x in np.sort(spec_x.unique())])

if len(mel_test) > 0: plt.scatter(mel_x, mel_test, label="Mel", color=colours['mel'])
if len(mfcc_test) > 0: plt.scatter(mfcc_x, mfcc_test, label="MFCC", color=colours['mfcc'])
if len(cqt_test) > 0: plt.scatter(cqt_x, cqt_test, label="CQT", color=colours['cqt'])
if len(spec_test) > 0: plt.scatter(spec_x, spec_test, label="Spectrogram", color=colours['spectrogram'])

if len(mel_test) > 0: plt.plot(mean_mel[:, 0], mean_mel[:, 1], color=colours['mel'], label='Mean Values')
if len(mfcc_test) > 0: plt.plot(mean_mfcc[:, 0], mean_mfcc[:, 1], color=colours['mfcc'], label='Mean Values')
if len(cqt_test) > 0: plt.plot(mean_cqt[:, 0], mean_cqt[:, 1], color=colours['cqt'], label='Mean Values')
if len(spec_test) > 0: plt.plot(mean_spec[:, 0], mean_spec[:, 1], color=colours['spectrogram'], label='Mean Values')

plt.ylim(0, 100)
plt.xscale('log')
plt.legend()
plt.show()
```
