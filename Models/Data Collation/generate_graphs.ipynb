```python
import pandas as pd
from matplotlib import pyplot as plt
import tikzplotlib
import os
```

# DIRECT ENCODING (Train and Test): Loss, Accuracy, F1, and AUC-ROC

```python
FILEPATH = "../IST non-JNB results/input_encoding/direct_encoding"
```

```python
mel_train_results = pd.read_csv(f"{FILEPATH}/mel_spectrograms/train.csv")
mel_test_results = pd.read_csv(f"{FILEPATH}/mel_spectrograms/test.csv")

mfcc_train_results = pd.read_csv(f"{FILEPATH}/mfcc_spectrograms/train.csv")
mfcc_test_results = pd.read_csv(f"{FILEPATH}/mfcc_spectrograms/test.csv")

epochs = mel_train_results['Epoch']
```

```python
# Loss
plt.figure(figsize=(8, 6))
plt.plot(epochs, mel_train_results['Test Loss'], label="Mel Train",  color='red')
plt.plot(epochs, mel_test_results['Test Loss'], label="Mel Test", color='red')
plt.plot(epochs, mfcc_train_results['Test Loss'], label="MFCC Train",  color='blue')
plt.plot(epochs, mfcc_test_results['Test Loss'], label="MFCC Test", color='blue')

plt.xlabel('Epoch')
plt.ylabel(r'Loss $\mathcal{L}$')
plt.title('Direct Encoding Loss')

plt.grid(True)
tikzplotlib.save("graphs/direct_loss.tex")
```

```python
# Accuracy
plt.figure(figsize=(8, 6))
plt.plot(epochs, mel_train_results['Accuracy'], label="Mel Train",  color='red')
plt.plot(epochs, mel_test_results['Accuracy'], label="Mel Test", color='red')
plt.plot(epochs, mfcc_train_results['Accuracy'], label="MFCC Train",  color='blue')
plt.plot(epochs, mfcc_test_results['Accuracy'], label="MFCC Test", color='blue')

plt.xlabel('Epoch')
plt.ylabel(r'Accuracy (\%)')
plt.title('Direct Encoding Accuracy')

plt.grid(True)
tikzplotlib.save("graphs/direct_accuracy.tex")
```

```python
# F1
plt.figure(figsize=(8, 6))
plt.plot(epochs, mel_train_results['F1'], label="Mel Train",  color='red')
plt.plot(epochs, mel_test_results['F1'], label="Mel Test", color='red')
plt.plot(epochs, mfcc_train_results['F1'], label="MFCC Train",  color='blue')
plt.plot(epochs, mfcc_test_results['F1'], label="MFCC Test", color='blue')

plt.xlabel('Epoch')
plt.ylabel(r'F1 Score (\%)')
plt.title('Direct Encoding F1 Score')

plt.grid(True)
tikzplotlib.save("graphs/direct_f1.tex")
```

```python
# AUC-ROC
plt.figure(figsize=(8, 6))
plt.plot(epochs, mel_train_results['AUC-ROC'], label="Mel Train",  color='red')
plt.plot(epochs, mel_test_results['AUC-ROC'], label="Mel Test", color='red')
plt.plot(epochs, mfcc_train_results['AUC-ROC'], label="MFCC Train",  color='blue')
plt.plot(epochs, mfcc_test_results['AUC-ROC'], label="MFCC Test", color='blue')

plt.xlabel('Epoch')
plt.ylabel(r'AUC-ROC (\%)')
plt.title('Direct Encoding AUC-ROC')

plt.grid(True)
tikzplotlib.save("graphs/direct_aucroc.tex")
```

# Time Contrast (Threshold) No Off-Spikes: F1 and AUC-ROC (single graph)

```python
max_step_forward = pd.read_csv(f"max_step_forward.csv")

max_thresholds = pd.read_csv(f"max_thresholds.csv")

mel_off_spikes_SF = max_step_forward[(max_step_forward['Off Spike'] == True) & (max_step_forward['Spectrogram'] == 'mel_spectrograms')]
mel_off_spikes_Th = max_thresholds[(max_thresholds['Off Spike'] == True) & (max_thresholds['Spectrogram'] == 'mel_spectrograms')]

mel_no_off_spikes_SF = max_step_forward[(max_step_forward['Off Spike'] == False) & (max_step_forward['Spectrogram'] == 'mel_spectrograms')]
mel_no_off_spikes_Th = max_thresholds[(max_thresholds['Off Spike'] == False) & (max_thresholds['Spectrogram'] == 'mel_spectrograms')]

mfcc_off_spikes_SF = max_step_forward[(max_step_forward['Off Spike'] == True) & (max_step_forward['Spectrogram'] == 'mfcc_spectrograms')]
mfcc_off_spikes_Th = max_thresholds[(max_thresholds['Off Spike'] == True) & (max_thresholds['Spectrogram'] == 'mfcc_spectrograms')]

mfcc_no_off_spikes_SF = max_step_forward[(max_step_forward['Off Spike'] == False) & (max_step_forward['Spectrogram'] == 'mfcc_spectrograms')]
mfcc_no_off_spikes_Th = max_thresholds[(max_thresholds['Off Spike'] == False) & (max_thresholds['Spectrogram'] == 'mfcc_spectrograms')]


thresholds = [0.01, 0.025, 0.05, 0.10, 0.20, 0.50]

# Train  AUC-ROC
plt.figure(figsize=(8, 6))
plt.plot(thresholds, mel_off_spikes_SF['Max AUC-ROC'], label="SF Mel Off Spikes",  color='DarkRed')
plt.plot(thresholds, mel_off_spikes_Th['Max AUC-ROC'], label="Th Mel Off Spikes", color='red')
plt.plot(thresholds, mel_no_off_spikes_SF['Max AUC-ROC'], label="SF Mel No Off Spikes",  color='DarkRed')
plt.plot(thresholds, mel_no_off_spikes_Th['Max AUC-ROC'], label="Th Mel No Off Spikes", color='red')
plt.plot(thresholds, mfcc_off_spikes_SF['Max AUC-ROC'], label="SF MFCC Off Spikes",  color='Navy')
plt.plot(thresholds, mfcc_off_spikes_Th['Max AUC-ROC'], label="Th MFCC Off Spikes", color='SkyBlue')
plt.plot(thresholds, mfcc_no_off_spikes_SF['Max AUC-ROC'], label="SF MFCC No Off Spikes",  color='Navy')
plt.plot(thresholds, mfcc_no_off_spikes_Th['Max AUC-ROC'], label="Th MFCC No Off Spikes", color='SkyBlue')


plt.xscale('log')
plt.xlabel(r'Threshold (\% of data)')
plt.ylabel(r'AUC-ROC (\%)')
plt.title('Naive Time Contrast Test AUC-ROC')

plt.grid(True)
tikzplotlib.save("graphs/tc_test_aucroc.tex")
```

# MLP-Beta

```python
FILEPATH = "../IST non-JNB results/architectures/MLP"

test_7 = pd.read_csv(f"{FILEPATH}/0.7/test.csv")
test_8 = pd.read_csv(f"{FILEPATH}/0.8/test.csv")
test_9 = pd.read_csv(f"{FILEPATH}/0.9/test.csv")
test_95 = pd.read_csv(f"{FILEPATH}/0.95/test.csv")

epochs = test_7['Epoch']

plt.figure(figsize=(8, 6))

plt.plot(epochs, test_7['F1'], label="0.70")
plt.plot(epochs, test_8['F1'], label="0.80")
plt.plot(epochs, test_9['F1'], label="0.90")
plt.plot(epochs, test_95['F1'], label="0.95")

plt.xlabel(r'Epoch')
plt.ylabel(r'F1 (\%)')
plt.title(r'MLP Comparison of $\beta$ in Test F1 Score')
plt.legend()

plt.grid(True)
```
