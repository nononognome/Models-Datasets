```python
import pandas as pd
from matplotlib import pyplot as plt
import tikzplotlib
import os
```

# DIRECT ENCODING (Train and Test): Loss, Accuracy, F1, and AUC-ROC

```python
FILEPATH = "../IST non-JNB results/input_encoding/direct_encoding"
```

```python
mel_train_results = pd.read_csv(f"{FILEPATH}/mel_spectrograms/train.csv")
mel_test_results = pd.read_csv(f"{FILEPATH}/mel_spectrograms/test.csv")

mfcc_train_results = pd.read_csv(f"{FILEPATH}/mfcc_spectrograms/train.csv")
mfcc_test_results = pd.read_csv(f"{FILEPATH}/mfcc_spectrograms/test.csv")

epochs = mel_train_results['Epoch']
```

```python
# Loss
plt.figure(figsize=(8, 6))
plt.plot(epochs, mel_train_results['Test Loss'], label="Mel Train",  color='red')
plt.plot(epochs, mel_test_results['Test Loss'], label="Mel Test", color='red')
plt.plot(epochs, mfcc_train_results['Test Loss'], label="MFCC Train",  color='blue')
plt.plot(epochs, mfcc_test_results['Test Loss'], label="MFCC Test", color='blue')

plt.xlabel('Epoch')
plt.ylabel(r'Loss $\mathcal{L}$')
plt.title('Direct Encoding Loss')

plt.grid(True)
tikzplotlib.save("graphs/direct_loss.tex")
```

```python
# Accuracy
plt.figure(figsize=(8, 6))
plt.plot(epochs, mel_train_results['Accuracy'], label="Mel Train",  color='red')
plt.plot(epochs, mel_test_results['Accuracy'], label="Mel Test", color='red')
plt.plot(epochs, mfcc_train_results['Accuracy'], label="MFCC Train",  color='blue')
plt.plot(epochs, mfcc_test_results['Accuracy'], label="MFCC Test", color='blue')

plt.xlabel('Epoch')
plt.ylabel(r'Accuracy (\%)')
plt.title('Direct Encoding Accuracy')

plt.grid(True)
tikzplotlib.save("graphs/direct_accuracy.tex")
```

```python
# F1
plt.figure(figsize=(8, 6))
plt.plot(epochs, mel_train_results['F1'], label="Mel Train",  color='red')
plt.plot(epochs, mel_test_results['F1'], label="Mel Test", color='red')
plt.plot(epochs, mfcc_train_results['F1'], label="MFCC Train",  color='blue')
plt.plot(epochs, mfcc_test_results['F1'], label="MFCC Test", color='blue')

plt.xlabel('Epoch')
plt.ylabel(r'F1 Score (\%)')
plt.title('Direct Encoding F1 Score')

plt.grid(True)
tikzplotlib.save("graphs/direct_f1.tex")
```

```python
# AUC-ROC
plt.figure(figsize=(8, 6))
plt.plot(epochs, mel_train_results['AUC-ROC'], label="Mel Train",  color='red')
plt.plot(epochs, mel_test_results['AUC-ROC'], label="Mel Test", color='red')
plt.plot(epochs, mfcc_train_results['AUC-ROC'], label="MFCC Train",  color='blue')
plt.plot(epochs, mfcc_test_results['AUC-ROC'], label="MFCC Test", color='blue')

plt.xlabel('Epoch')
plt.ylabel(r'AUC-ROC (\%)')
plt.title('Direct Encoding AUC-ROC')

plt.grid(True)
tikzplotlib.save("graphs/direct_aucroc.tex")
```
